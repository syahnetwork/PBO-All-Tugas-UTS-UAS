## Class dan Object

### Pengenalan
Paradigma baru yang muncul sejak tahun 1990-an bagi bahasa pemrograman untuk menggunakan konstruksi khusus yang disebut class dan object. Bahasan pemrograman ini yang saat ini desebut sebagai Object Oriented Programming atau Pemrograman Berorientasi Objek.

Pada bab ini dan yang berikutnya kita akan menjelaskan apa tentang paradigma PBO, dan menunjukkan bagaimana kita dapat memanfaatkan kekuatan penuh dari bahasa berorientasi objek seperti Java.

### Class Sebagai Tipe Data

Sejauh ini Anda telah menggunakan tipe data seperti int, float dan char. Ini adalah tipe data sederhana / dasar yang menyimpan satu informasi. Tetapi bagaimana jika kita ingin variabel untuk menyimpan lebih dari satu informasi yang terkait/
Sebagai contoh misalnya sebuah buku — sebuah buku mungkin memiliki a judul, penerbit dan ISBN. Atau Mahasiswa yang memiliki atribut ID, nama dan alamat.

Bahasa berorientasi objek seperti Java dan C ++ mempunyai kemampuan lebih jauh. Mereka memungkinkan kita tidak hanya mendefinisikan variable dan tipe data yang akan menyimpan banyak data, tetapi juga bisa mendefinisikan  metode /fungsi dalam sebuah tipe data class, dimana method terebut bisa kita gunakan untuk memproses data itu.

Sebagai contoh Class Buku selain atribut judul, penerbit, ISBN tetapi juga memiliki fungsi/method untuk menghitung diskon dan pajak pembelian. Kemudian Class Mahasiswa atribut ID, nama dan alamat memiliki method untuk menghitung Nilai IPK

### [3] Kompunen Konstruktor

Konstruktor adalah merupakan fungsi atau method dari suatu class tetapi memiliki ciri khusus, yaitu namanya harus sama dengan class

Lakukan eksekusi dari program *Program 5.11* dan *Program 5.12* berikan penjelasan anda mengenai dua program tersebut

````scijava
//Program 5.11 Class Calculator6 dengan konstrukor

class Calculator6{
    int x;
    int y;
    int hasil;
    
  Calculator6(int varX, int varY){
    x = varX;  
    y = varY;  
  }
    
  void doTambah(){
     System.out.println("Operasi Tambah");  
     hasil = x + y;
  }   

  void prosedurHitung(char simbol){
     if(simbol == '+'){
         
        this.doTambah();
         
     }      
  }  
   
  int fungsiHitung(char simbol){
     if(simbol == '+'){
         
        this.doTambah();
         
     }
     return hasil; 
  }    
}
````

````scijava
//Program 5.12 Implementasi Class Calculator6 dengan konstrukor
Calculator6 oCalculator = new Calculator6(10,20);
System.out.println("Hasil "+oCalculator.fungsiHitung('+')); 
````
